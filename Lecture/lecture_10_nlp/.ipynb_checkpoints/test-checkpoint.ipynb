```python
import numpy as np
```

```python
n_components = 2
p=[0]*n_components
x =np.array([-5,-4,-3,-2, 3,4,5]).reshape((7,1))
pi = np.array([0.8, 0.2])
mu = np.array([-1, 1])
sigma = np.array([1,1])
print('mu = ' , mu)
print('sigma = ', sigma)
print('pi = ', pi)
print('x = ', x)
print('p = ', p)
shape_0 = len(x)
shape_1 = 1
for i in range(10):
    for k in range(n_components):
        #pdf - probability density function
        #print(self.mu[k], self.sigma[k])
        p[k] = pi[k] * multivariate_normal.pdf(x, mean = mu[k], cov=sigma[k])
    p = p/(np.sum(p, axis=0))
    # N_k
    N_k = np.sum(p, axis=1)

    mu = (p @ x)/N_k.reshape(-1,1)
    for k in range(n_components):
        elem = (x - mu[k])
        print(elem)
        sigma[k] = sum(p[k]*(elem.T @ elem)[0])/N_k[k]/104
    pi = N_k/shape_0 
    print("-"*20, i,"-"*20)
    print('mu = ' , mu)
    print('sigma = ', sigma)
    print('pi = ', pi)
    print('x = ', x)
    print('p = ', p)
```
